```python
import pandas as pd
import matplotlib.pyplot as plt
#first import packages into python
```

```python
data=pd.read_csv('countries.csv')
```

```python
#create filter to create the traget countries
china_filter=data.country=='China'
######use bolean filter
US_filter=data.country=='United States'
Japan_filter=data.country=='Japan'
Indian_filter=data.country =='India'
```

```python
####create filter to convience next step analysis
china_data=data[china_filter]
US_data=data[US_filter]
Japan_data=data[Japan_filter]
Indian_data=data[Indian_filter]
```

```python
#create plot 
plt.plot(china_data.year, china_data.population/china_data.population.iloc[0])
plt.plot(US_data.year,US_data.population/US_data.population.iloc[0])
plt.plot(Japan_data.year,Japan_data.population/Japan_data.population.iloc[0])
plt.plot(Indian_data.year, Indian_data.population/Indian_data.population.iloc[0])
plt.title('country population')
plt.ylabel('population growth')
plt.xlabel('year')
plt.legend(['China','US','Japan','Indian'])
plt.show()
```
